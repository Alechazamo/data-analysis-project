```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
hypotheses = pd.read_csv('/datasets/hypotheses_us.csv', sep = ';')
```

```python
pd.set_option('display.max_colwidth', None)
print (hypotheses)
```

```python
hypotheses['ICE'] = hypotheses['Impact'] * hypotheses['Confidence'] / hypotheses['Effort']
ice_sorted = hypotheses.sort_values(by = 'ICE', ascending=False)
print(ice_sorted[['Hypothesis', 'ICE']])
```

Esta tabla muestra que la hipótesis #8 tiene el ICE más alto 'Launch a promotion that gives users discounts on their birthdays'. El score es de 16.2. Esto significa que considerando el impacto, la confianza y el esfuerzo, esta hipótesis tene la mejor relación costo-beneficio para aumentar los ingresos.

```python
hypotheses['RICE'] = (hypotheses['Impact'] * hypotheses['Reach'] * hypotheses['Confidence']) / hypotheses['Effort']
print(hypotheses[['Hypothesis', 'RICE']])
```

```python
rice_sorted= hypotheses.sort_values(by= 'RICE', ascending=False)
print(rice_sorted[['Hypothesis', 'RICE']])
```

Comparación de ICE vs RICE

Cuando aplicamos el método ICE, la hipótesis con mayor prioridad fue "Lanzar una promoción que da descuentos en cumpleaños". Sin embargo, al aplicar RICE, la hipótesis principal cambió a "Agregar un formulario de suscripción en todas las páginas principales".

Esto ocurre porque RICE toma en cuenta el alcance. La hipótesis del formulario afecta a más usuarios, por lo que su prioridad aumentó significativamente.

En cambio, otras hipótesis que tenían buen impacto pero poco alcance bajaron posiciones con RICE.

En resumen, al usar RICE, las hipótesis que benefician a más usuarios son priorizadas, lo que nos ayuda a elegir acciones con mayor efecto global.

```python
orders = pd.read_csv('/datasets/orders_us.csv')
orders['date'] = pd.to_datetime(orders['date'])

visits = pd.read_csv('/datasets/visits_us.csv')
visits['date'] = pd.to_datetime(visits['date'])


print(orders.head())
print(visits.head())
```

```python
datesGroups = orders[['date', 'group']].drop_duplicates()

ordersAggregated = datesGroups.apply(
    lambda x: orders[
        (orders['date'] <= x['date']) & (orders['group'] == x['group'])
    ].agg({
        'date': 'max',
        'group': 'max',
        'transactionId': 'nunique',
        'visitorId': 'nunique',
        'revenue': 'sum'
    }),
    axis=1
).sort_values(by=['date', 'group'])
```

```python
plt.figure(figsize=(12,6))
for grupo in ['A', 'B']:
    data = ordersAggregated[ordersAggregated['group'] == grupo]
    plt.plot(data['date'], data['revenue'], label=f'Grupo {grupo}')

plt.legend()
plt.title('Ingresos Acumulados por Grupo')
plt.xlabel('Fecha')
plt.ylabel('Ingresos Acumulados')
plt.show()
```

```python
cumulativeData = ordersAggregated.copy()

cumulativeRevenueA = cumulativeData[cumulativeData['group'] == 'A'][['date', 'revenue', 'transactionId']]
cumulativeRevenueB = cumulativeData[cumulativeData['group'] == 'B'][['date', 'revenue', 'transactionId']]
```

```python
plt.figure(figsize=(12,6))

plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'] / cumulativeRevenueA['transactionId'], label='Grupo A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'] / cumulativeRevenueB['transactionId'], label='Grupo B')

plt.legend()
plt.title('Tamaño Promedio de Pedido Acumulado por Grupo')
plt.xlabel('Fecha')
plt.ylabel('Tamaño Promedio de Pedido')
plt.show()
```

Grupo A (línea azul):

Se mantiene bastante estable después de unos días
El tamaño promedio de pedido ronda entre 100 y 120

Grupo B (línea naranja):

Tiene una gran subida alrededor del 17 de agosto
Después de esa subida, el tamaño promedio empieza a bajar ligeramente pero sigue siendo mayor que el del Grupo A

Parece que hubo un pedido muy grande o varios pedidos altos en el Grupo B, lo que generó ese pico, esto podría ser una anmalía o un outlier

Pero hasta ahora, puedi pensar que el Grupo B tiene un tamaño promedio mayor que el Grupo A


```python
mergedCumulativeRevenue = cumulativeRevenueA.merge(
    cumulativeRevenueB,
    on='date',
    suffixes=['A', 'B']
)

plt.figure(figsize=(12,6))
plt.plot(
    mergedCumulativeRevenue['date'],
    (mergedCumulativeRevenue['revenueB'] / mergedCumulativeRevenue['transactionIdB']) /
    (mergedCumulativeRevenue['revenueA'] / mergedCumulativeRevenue['transactionIdA']) - 1
)

plt.axhline(y=0, color='black', linestyle='--')
plt.title('Diferencia Relativa del Tamaño Promedio de Pedido: B vs A')
plt.xlabel('Fecha')
plt.ylabel('Diferencia Relativa')
plt.show()
```

Al inicio la fluctuación del Grupo B es muy inestable, teniendo muchos picos y caídas y cruzando varias veces cero. A la mitad del periodo, hay un aumento abrupto en la gráfica por parte del grupo B, esto puede deberse a alguna anomalía, como a que se vió reflejada anteriormente y puede deberse a un outlier. Sin embargo, posteriormente, la línea baja un poco pero se mantiene estable por encima del A considerablemtne.

Más adelante, al eliminar anomalías y hacer pruebas estadísticas, confirmaremos si esta diferencia es significativa o solo producto del azar.

```python
orders_by_day = orders.groupby(['date', 'group']).agg({'transactionId': 'nunique'}).reset_index()
conversion_data = orders_by_day.merge(visits, on = ['date', 'group'])
conversion_data['conversion'] = conversion_data['transactionId'] / conversion_data['visits']
```

```python
plt.figure(figsize=(12,6))
for grupo in ['A', 'B']:
    data = conversion_data[conversion_data['group'] == grupo]
    plt.plot(data['date'], data['conversion'], label=f'Grupo {grupo}')

plt.legend()
plt.title('Tasa de Conversión Diaria por Grupo')
plt.xlabel('Fecha')
plt.ylabel('Tasa de Conversión')
plt.show()
```

La gráfica muestra una alta volatilidad en las tasas de conversión diarias tanto para el grupo A como para el grupo B a lo largo del periodo analizado. Se observan múltiples picos y caídas abruptas, lo que indica que las conversiones diarias no fueron consistentes.

A pesar de esta inestabilidad, hacia el final del periodo, el grupo B presenta una tendencia a mantener tasas de conversión ligeramente superiores en comparación con el grupo A.

```python
orders_by_users_A = orders[orders['group'] == 'A'].groupby('visitorId', as_index=False).agg({'transactionId': 'nunique'})
orders_by_users_A.columns = ['visitorId', 'orders']

orders_by_users_B = orders[orders['group'] == 'B'].groupby('visitorId', as_index=False).agg({'transactionId': 'nunique'})
orders_by_users_B.columns = ['visitorId', 'orders']

sampleA = pd.concat([
    orders_by_users_A['orders'],
    pd.Series(0, index=np.arange(visits[visits['group'] == 'A']['visits'].sum() - len(orders_by_users_A)))
], axis=0)

sampleB = pd.concat([
    orders_by_users_B['orders'],
    pd.Series(0, index=np.arange(visits[visits['group'] == 'B']['visits'].sum() - len(orders_by_users_B)))
], axis=0)
```

```python
u_statistic, p_value = stats.mannwhitneyu(sampleA, sampleB)
print(f"Valor p: {p_value}")
```

Esto significa que la diferencia entre el grupo A y el grupo B es estadísticamente significativa. Podemos rechazar la hipotesis nula y concluir que el grupo B está teniendo un performing diferente

```python
relative_diff = sampleB.mean() / sampleA.mean() - 1
print(f"Diferencia relativa: {relative_diff:.2%}")
```

Tras realizar la prueba de significancia estadística para comparar las tasas de conversión entre los grupos A y B, obtuve un valor p = 0.0167, que es menor al común de 0.05. Esto indica que la diferencia observada entre ambos grupos es estadísticamente significativa.

Además, el grupo B presenta una diferencia de 13.81%, lo que significa que convierte aproximadamente un 14% más que el grupo A.

Por lo tanto, considero prudente detener la prueba A/B y considerar al grupo B como el líder, implementando las estrategias aplicadas en este grupo para maximizar los ingresos y la conversión general del negocio.
